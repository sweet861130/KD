```python
excelname= "kd"
sheetname = "kd"

wb = openpyxl.load_workbook(str(excelname)+".xlsx")
ws = wb[str(sheetname)]

dataset = []

i = 1
while(ws.cell(row=i, column=1).value!=None):
    row = []
    j = 1
    while(ws.cell(row=i, column=j).value!=None):
        row.append(ws.cell(row=i, column=j).value)
        j = j +1
    dataset.append(row)
    i = i + 1
pd.DataFrame(dataset)
```

```python

# trade =[[0 for x in range(12)] for y in range(len(dataset))]
trade = []
for i in range(0, len(dataset)):
    row = []
    for j in range(0, 12):
        row.append(0)
    trade.append(row)
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "rsv"
trade[0][6] = "k"
trade[0][7] = "d"
trade[0][8] = "buy_day"
trade[0][9] = "cum_buy_day"
trade[0][10] = "sell_day"
trade[0][11] = "10ma"
pd.DataFrame(trade)
```

```python
for i in range(1, len(trade)):#iå¾ž1åˆ°(è³‡æ–™åˆ—æ•¸)çš„è¿´åœˆ
    trade[i][0] = dataset[i][0]#tradeçš„ç¬¬nåˆ—ç¬¬0è¡Œ=datasetè³‡æ–™é›†è£¡çš„ç¬¬nåˆ—ç¬¬0è¡Œ
    trade[i][4] = float(dataset[i][4])
for i in range(10, len(trade)):
    temp = 0
    for j in range(0, 10):
        temp = temp + trade[i-j][4]
    trade[i][11] = temp / 10
for i in range(9, len(trade)):
    lowest = []
    highest = []
    for j in range(0, 9):
        lowest.append(trade[i-j][3])
        highest.append(trade[i-j][2])
    trade[i][5] = (trade[i][4]-min(lowest))/(max(highest)-min(lowest))*100
trade[9][6] = 50
trade[9][7] = 50
for t in range(10, len(trade)):
    trade[t][6] = trade[t-1][6]*2/3 + trade[t][5]*1/3
    trade[t][7] = trade[t-1][7]*2/3 + trade[t][6]*1/3


pd.DataFrame(trade)
```

```python
for i in range(10, len(trade)):
    if (trade[i][6] > trade[i][7])and(trade[i-1][9]==0):
        trade[i][8] = 1#è²·
        trade[i][9] = 1#æŒæœ‰
        trade[i][10] = 0#ä¸è³£å‡º
    elif(trade[i][6] < trade[i][7])and(trade[i-1][9]==1):
        trade[i][8] = 0#ä¸è²·
        trade[i][9] = 0#æ²’æŒæœ‰
        trade[i][10] = 1#è³£å‡º
    elif(trade[i-1][9]>0):
        trade[i][8] = 0#ä¸è²·
        trade[i][9] = 1#æŒæœ‰
        trade[i][10] = 0#ä¸è³£å‡º
pd.DataFrame(trade)
```
